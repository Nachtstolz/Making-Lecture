if ~ else 조건문 문법 이해하기 -> ppt 참고

```python
# 점수에 따른 장학금 대상자 여부 알려주기
score = int(input("점수가 몇 점인가요? "))
if score >= 90 :
    print("장학금 대상자입니다.")
    print("축하합니다.")
else :
    print("장학금 대상자가 아닙니다.")
    print("다음 학기를 노려봅시다.")
print("수고하셨습니다.")
```

여러 가지 조건을 지정, 조건에 맞는 명령문을 수행하기 위한 if ~ elif ~ else 문 -> ppt 참고

```python
score = int(input("시험 점수를 입력하세요: "))
if score >= 90 :
    print("시험을 아주 잘 봤군요. 축하해요.")
elif score>=80 :
    print("시험을 괜찮게 봤군요. 수고했어요.")
elif score>=70 :
    print("시험을 좀 못봤군요. 다음에는 잘 봐요.")
else :
    print("완전히 망했군요.")
```

### 반복문 활용하기

반복문 문법 익히기 -> PPT 참고

**while**

```python
# while 반복문으로 숫자 세기
n = 1
while n <= 10 :
    print(n)
    n = n+1
print('숫자 세기 끝')
```

```python
# 1초 간격으로 숫자 출력되도록 보완
import time
n = 1
while n <= 10 :
    print(n)
    n = n+1
    time.sleep(1)
print('숫자 세기 끝')
```

```python
# Count 1, Count 2 ... 형태로 1부터 10까지 세는 프로그램
n = 1
while n<=10 :
    print("Count ", n)
    n +=1
print("종료합니다.")
```

**for**

```python
# 10부터 1까지 거꾸로 세기
for n in range(10, 0, -1) :
    print("Count %d" %n)
print("종료합니다")
```

```python
# for 반복문의 파라미터 생략하기
# 세 번째 파라미터 : 증가값 생략하기
for n in range(1, 11) :
    print("Count %d" %n)
print("종료합니다.")
```

```python
# 첫 번째, 세 번째 파라미터 : 초기값, 증가값 생략하기
# 위 코드를 여기에 복사 및 수정하여 실행해보자
for n in range(11) :
    print("Count %d" %n)
print("종료합니다.")
```

```python
# 특정 횟수만큼 반복하고 싶다면
# for :
for n in range(5) :
    print('Hello Python')
print('end')
```

```python
# while
n = 1
while n<=5 :
    print('Hello Python')
    n = n+1
print('end')
```

```python
# 무한 반복하기
while True : # while 1 : 과 동일
    print('Hello Python')
print('end')
```

```python
while 0 :
    print('Hello Python')
print('end')
```

```python
# 무한 반복에서 탈출하기
while 1 :
    year = int(input('몇 년도에 태어나셨나요? '))
    if year == 0 :
        break # 브레이크 문을 통해 빠져나오기
    age = 2022 - year + 1
    print('그럼 올해 %d살 이겠군요\n' %age)
    
print('프로그램을 사용해주셔서 감사합니다.')
```
